```
import math
import matplotlib.pyplot as plt

class color:
   RED = '\033[91m'  
   END = '\033[0m'
   UNDERLINE = '\033[4m'

# Initial positions at t=0
x_A, y_A = [10], [0]
x_B, y_B = [0],  [10]
x_C, y_C = [10], [10]
x_D, y_D = [0],  [0]

# Velocities
v_A, v_B, v_C, v_D = 3, 5, 7, 2

# Distances
d_AD, d_CA, d_BC, d_DB = [], [], [], []

# Total  
A = B = C = D = 0

for t in range(0,21):
  
  print("\n",color.UNDERLINE,"At time = ",t,color.END)
  print("x_A = ",x_A[t],"  y_A = ",y_A[t])
  print("x_B = ",x_B[t],"  y_B = ",y_B[t])
  print("x_C = ",x_C[t],"  y_C = ",y_C[t])
  print("x_D = ",x_D[t],"  y_D = ",y_D[t])

  d_DB.append(math.sqrt((x_D[t]-x_B[t])**2 + (y_D[t]-y_B[t])**2))
  print("D to B distance = ",d_DB[t])
  if d_DB[t] < 5:
    print(color.RED,"Car B shoots Car D at time ",t,color.END)
    D = D + 1
  sin = (y_D[t] - y_B[t])/d_DB[t]
  cos = (x_D[t] - x_B[t])/d_DB[t]
  x_B.append(x_B[t] + (v_B*cos))
  y_B.append(y_B[t] + (v_B*sin))

  d_CA.append(math.sqrt((x_C[t]-x_A[t])**2 + (y_C[t]-y_A[t])**2))
  print("C to A distance = ",d_CA[t])
  if d_CA[t] < 5:
    print(color.RED,"Car A shoots Car C at time ",t,color.END)
    C = C + 1
  sin = (y_C[t] - y_A[t])/d_CA[t]
  cos = (x_C[t] - x_A[t])/d_CA[t]
  x_A.append(x_A[t] + (v_A*cos))
  y_A.append(y_A[t] + (v_A*sin))

  d_BC.append(math.sqrt((x_B[t]-x_C[t])**2 + (y_B[t]-y_C[t])**2))
  print("B to C distance = ",d_BC[t])
  if d_BC[t] < 5:
    print(color.RED,"Car C shoots Car B at time ",t,color.END)
    B = B + 1
  sin = (y_B[t] - y_C[t])/d_BC[t]
  cos = (x_B[t] - x_C[t])/d_BC[t]
  x_C.append(x_C[t] + (v_C*cos))
  y_C.append(y_C[t] + (v_C*sin))

  d_AD.append(math.sqrt((x_A[t]-x_D[t])**2 + (y_A[t]-y_D[t])**2))
  print("A to D distance = ",d_AD[t])
  if d_AD[t] < 5:
    print(color.RED,"Car D shoots Car A at time ",t,color.END)
    A = A + 1
  sin = (y_A[t] - y_D[t])/d_AD[t]
  cos = (x_A[t] - x_D[t])/d_AD[t]
  x_D.append(x_D[t] + (v_D*cos))
  y_D.append(y_D[t] + (v_D*sin))
  

print("\n"+color.UNDERLINE+"Finally:"+color.END)
print("\tA Got Shoot ", A, "times")
print("\tB Got Shoot ", B, "times")
print("\tC Got Shoot ", C, "times")
print("\tD Got Shoot ", D, "times")

print("\n"+color.UNDERLINE+"Graph\n"+color.END)
plt.plot(x_A, y_A, color = "Blue", label = "Path A")
plt.plot(x_B, y_B, color = "Orange", label = "Path B")
plt.plot(x_C, y_C, color = "Green", label = "Path C")
plt.plot(x_B, y_B, color = "Red", label = "Path D")
plt.legend()
plt.show()
```
